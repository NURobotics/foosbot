```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Create a white 300x200 image
clean_img = np.zeros((300, 200, 3), np.uint8)
clean_img.fill(255)

# Draw a solid red box at position (50, 50) with width 100 and height 100
cv2.rectangle(clean_img, (50, 50), (150, 150), (255, 0, 0), -1)

pass
```

```python
# display the image
plt.imshow(clean_img)
plt.show()
```

```python
# Generate a transformation matrix to warp, rotate, and scale the image

transformation_matrix = np.float32([[1, 0, 0], [0, 1, 0]])

# Use cv2.warpAffine to apply the transformation matrix to the image

dirty_img = cv2.warpAffine(clean_img, transformation_matrix, (200, 300))
```

```python
# Display the image
plt.imshow(dirty_img)
plt.show()
```
